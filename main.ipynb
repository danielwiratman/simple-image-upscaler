## Image Resolution Enhancer
Made for UAS Ilmu Data II

```python
import numpy as np
from PIL import Image
from ISR.models import RRDN, RDN
```

```python
# model = RRDN(weights='gans')
model = RDN(weights='noise-cancel')
```

```python
def processImage(url):
    img = Image.open(url)
    print('Original Image')
    display(img)
    print('Original Image (Resized 4x)')
    display(img.resize((img.size[0]*2, img.size[1]*2)))
    sr_img = model.predict(np.array(img))
    print("Enhanced (4x Resolution)")
    display(Image.fromarray(sr_img))
```

```python
processImage('lowresPikachu.jpg')
```

```python
processImage('lowresPond.jpeg')
```

```python
processImage('lowresBlueFlower.jpeg')
```
